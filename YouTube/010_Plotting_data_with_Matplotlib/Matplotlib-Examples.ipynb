```python
import matplotlib.pyplot as plt
import numpy as np
```

**Making a scatter plot**

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16,25]
```

```python
plt.scatter(x_values, y_values)
```

```python
plt.scatter(x_values, y_values, marker = '2', color = 'green')
```

**Making a line chart**

```python
plt.plot(x_values, y_values)
```

```python
def plot_segment(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    plt.plot([x1, x2], [y1, y2], marker = '*', c = 'green')
```

```python
point1 = (0,3)
point2 = (2,1)
plot_segment(point1, point2)
```

```python
x_values = np.arange(0, 10, 0.01)
y_values = np.sin(x_values)
```

```python
plt.plot(x_values, y_values)
```

**More plot customizations**

```python
plt.ylim(0,5)
plt.xlim(0,5)
plt.gcf().set_size_inches(5,5)
plot_segment(point1, point2)
```

```python
x_values = np.arange(0, 10, 0.01)
y_values = np.sin(x_values)

plt.ylim(-5,5)
plt.xlim(0,10)
plt.grid(True)
plt.rcParams['grid.color'] = 'blue'


plt.plot(x_values, y_values)
plt.title('Graph of sin (x) vs. x', fontsize = 16)
plt.xlabel('x value', fontsize = 16)
plt.ylabel('The value of sin(x)', fontsize = 16)
```
