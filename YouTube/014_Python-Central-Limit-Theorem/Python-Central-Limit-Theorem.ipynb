```python
import random
import plotly.express as px
import matplotlib.pyplot as plt
import statistics
```

```python
sample_size = 31
sample_count = 10_000
```

```python
random_values = [random.uniform(3.14, 42) for _ in range(sample_size)]
```

```python
random_values.sort()
```

```python
def scatter_x_values(x):
    plt.figure(figsize=(10,6))
    plt.scatter(x, [1] * len(x), alpha=.3)
    plt.yticks([])
    plt.title(f'Scatter plot of {len(x)} values')
    plt.xlabel('Scattered values') #I have added that after the video :)
    plt.show()               
```

```python
scatter_x_values(random_values)
```

```python
statistics.mean(random_values)
```

```python
def generate_sample_average(sample_size):
    return sum(random.uniform(3.14, 42) for _ in range(sample_size)) / sample_size
```

```python
generate_sample_average(sample_size)
```

```python
x_values = [generate_sample_average(sample_size) for _ in range(sample_count)]
```

```python
y_values = [1] * sample_count
```

```python
scatter_x_values(x_values)
```

```python
px.histogram(x = x_values, y = y_values, nbins = 0).show()
```
