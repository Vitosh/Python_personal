```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml
```

```python
# Load MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
images = mnist.data.to_numpy().reshape(-1, 28, 28)
labels = mnist.target.astype(int)
```

```python
plt.imshow(images[11], cmap='gray')
plt.title(f'Digit {labels[11]}')
plt.show()
```

```python
eight_images = images[labels == 8]
sum_image = np.sum(eight_images, axis=0)

plt.imshow(sum_image, cmap='Greens')
plt.colorbar()
plt.title(f'Digit 8')
plt.show()
```

```python
three_images = images[labels == 3]
synthetic_digit = np.sum(three_images[:3], axis = 0)

plt.imshow(synthetic_digit, cmap='Greens')
plt.colorbar()
plt.title(f'Digit 3')
plt.show()
```

```python
def create_text_image(text, digits_data, labels, spacing = 2):
    letter_height = 28
    letter_width = 28
    result_image = np.zeros((letter_height, len(text) * (letter_width+spacing)))
    
    for idx, char in enumerate(text):
        digit = int(char) if char.isdigit() else 1
        char_image = digits_data[labels==digit][23]
        start_pos = idx * (letter_width + spacing)
        result_image[:, start_pos:start_pos + letter_width] = char_image
        
    return result_image

image_of_digits = "43321337"
my_image = create_text_image(image_of_digits, images, labels, 2)
plt.imshow(my_image, cmap='plasma')
plt.show()
```

```python
def create_text_image_multiple(text, digits_data, labels, repetitions, spacing = 0):
    letter_height = 28
    letter_width = 28
    result_image = np.zeros((letter_height, len(text) * (letter_width+spacing)))
    
    for idx, char in enumerate(text):
        digit = int(char) if char.isdigit() else 1
        char_image = digits_data[labels==digit][repetitions]
        start_pos = idx * (letter_width + spacing)
        result_image[:, start_pos:start_pos + letter_width] = char_image
        
    return result_image

image_of_digits = "1234567890"
for i in range(100, 111):
    my_image = create_text_image_multiple(image_of_digits, images, labels, i, 2)
    plt.imshow(my_image, cmap='plasma')
    plt.show()
```
