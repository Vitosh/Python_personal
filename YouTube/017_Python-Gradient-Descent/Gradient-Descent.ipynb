```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2
```

```python
x_vals = np.linspace(-100, 100, 1000)
y_vals = f(x_vals)
```

```python
plt.figure(figsize = (10,5))
plt.plot(x_vals, y_vals, label="f(x)=x^2", color = "green")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = x^2')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = 10
learning_rate = 0.01
iterations = 10111
iteration_data = []
history = [x]
```

```python
for i in range(iterations + 1):
    gradient = 2 * x
    x = x - learning_rate * gradient
    history.append(x)
```

```python
y_history = [f(x) for x in history]
```

```python
plt.figure(figsize = (10,5))
plt.plot(x_vals, y_vals, label="f(x)=x^2", color = "green")
plt.scatter(history, y_history, color='red',label = "Descent Path",zorder=5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Gradient Descent - first {len(y_history)} interations')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.xlim(-30, 30)
plt.ylim(0,1100)
plt.legend()
plt.grid(True)
plt.show()
```
