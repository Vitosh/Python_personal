```python
# Triangle coordinates for 30-degree angle
x = [0, 2, 0]
y = [0, 0, 1]

# Plot the triangle
plt.figure(figsize=(6, 6))
plt.plot(x + [0], y + [0], 'b-', linewidth=2)  # Triangle sides
plt.fill_between(x, y, alpha=0.1, color="blue")  # Shading the triangle

# Annotating the sides
plt.text(1, -0.1, "Adjacent = √3", fontsize=10, ha="center", va="top")
plt.text(-0.2, 0.5, "Opposite = 1", fontsize=10, rotation=90, ha="right", va="center")
plt.text(1, 0.3, "Hypotenuse = 2", fontsize=10, rotation=-30, va="bottom")

# Annotating the angle
plt.text(2, 0.1, r"$30^\circ$", fontsize=14)

# Ensuring equal aspect ratio for accurate angle representation
plt.axis("equal")
plt.xlim(-0.5, 2.5)
plt.ylim(-0.5, 1.5)
plt.axhline(0, color="gray", linewidth=0.5, linestyle="dotted")
plt.axvline(0, color="gray", linewidth=0.5, linestyle="dotted")
plt.title("Geometric Definition of Sine", fontsize=14)

plt.show()
```

```python
# This one is animated, if it is removed from the Jupyter Notebook and ran separately as a python file.
# That is why the imports are repated :)

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Generate unit circle points
theta = np.linspace(0, 2 * np.pi, 1000)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Initialize figure
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x_circle, y_circle, 'b-', label="Unit Circle")  # Unit circle
ax.axhline(0, color="gray", linestyle="dotted")
ax.axvline(0, color="gray", linestyle="dotted")

# Add dynamic triangle components
triangle_line, = ax.plot([], [], 'r-', linewidth=2, label="Triangle Sides")
point, = ax.plot([], [], 'ro')  # Moving point on the circle

# Text for dynamic values
dynamic_text = ax.text(0.03, 0.03, "", fontsize=12, color="black", ha="left", transform=ax.transAxes)

# Set up axis limits and labels
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_title("Sine as a Triangle on the Unit Circle", fontsize=14)
ax.set_xlabel("cos(θ)", fontsize=12)
ax.set_ylabel("sin(θ)", fontsize=12)
ax.legend(loc="upper left")

# Animation update function
def update(frame):
    angle = theta[frame]
    x_point = np.cos(angle)
    y_point = np.sin(angle)
    degrees = np.degrees(angle) % 360  # Convert radians to degrees
    
    # Update triangle
    triangle_line.set_data([0, x_point, x_point, 0], [0, y_point, 0, 0])
    
    # Update point on the circle
    point.set_data([x_point], [y_point])  # Fixed this line to avoid the warning
    
    # Update text for angle, opposite side length, and sin(θ)
    dynamic_text.set_text(f"Angle: {degrees:.1f}°\nOpposite Side Length: {y_point:.2f}\nsin(θ): {y_point:.2f}")
    return triangle_line, point, dynamic_text

# Create animation
ani = animation.FuncAnimation(fig, update, frames=len(theta), interval=20, blit=True)
plt.show()
```

```python
# Function to plot a triangle
def plot_triangle(angle_deg, hypotenuse, opposite, adjacent, angle_label):
    hypotenuse = round(hypotenuse, 2)
    opposite = round(opposite, 2)
    adjacent = round(adjacent, 2)
    
    plt.figure(figsize=(6, 6))
    x = [0, adjacent, 0]  # x-coordinates
    y = [0, 0, opposite]  # y-coordinates
    
    plt.plot(x + [0], y + [0], 'b-', linewidth=2)
    plt.fill_between(x, y, alpha=0.1, color="blue")
    
    plt.text(adjacent / 2, -0.2, f"Adjacent = {adjacent}", fontsize=10, ha="center")
    plt.text(-0.3, opposite / 2, f"Opposite = {opposite}", fontsize=10, rotation=90, ha="center")
    plt.text(adjacent / 2 + 0.5, opposite / 2, f"Hypotenuse = {hypotenuse}", fontsize=10, ha="center")
    plt.text(adjacent + 0.2, 0, angle_label, fontsize=14)
    
    plt.axis("equal")
    plt.xlim(-1, hypotenuse + 1)
    plt.ylim(-1, hypotenuse + 1)
    plt.axhline(0, color="gray", linestyle="dotted")
    plt.axvline(0, color="gray", linestyle="dotted")
    plt.title(f"Sine of {angle_deg}°", fontsize=14)
    plt.show()

# 30° Triangle
plot_triangle(30, 2, 1, np.sqrt(3), r"$30^\circ$")

# 45° Triangle
plot_triangle(45, np.sqrt(2), 1, 1, r"$45^\circ$")

# 60° Triangle
plot_triangle(60, 2, np.sqrt(3), 1, r"$60^\circ$")
```

```python
# Generate sound wave data
time = np.linspace(0, 2, 1000)  # Time in seconds
frequency = 5  # Frequency in Hz
amplitude = 2  # Amplitude

sound_wave = amplitude * np.sin(2 * np.pi * frequency * time)

# Plot the sound wave
plt.figure(figsize=(10, 6))
plt.plot(time, sound_wave, 'g-', label="Sound Wave (sine)", linewidth=2)
plt.title("Sound Wave Modeled as Sine", fontsize=14)
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
# Generate x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.sin(x)

# Plot the sine function
plt.figure(figsize=(10, 5))
plt.plot(x, y, label="sin(x)", linewidth=2)

# Customize x-axis with π labels
xticks = [-2 * np.pi, -3 * np.pi / 2, -np.pi, -np.pi / 2, 0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi]
xtick_labels = [r"$-2\pi$", r"$-\frac{3\pi}{2}$", r"$-\pi$", r"$-\frac{\pi}{2}$", r"$0$", 
                r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]
plt.xticks(xticks, xtick_labels)

# Add grid, title, and labels
plt.axhline(0, color="black", linewidth=0.8, linestyle="dotted")
plt.axvline(0, color="black", linewidth=0.8, linestyle="dotted")
plt.title("Sine Function", fontsize=16)
plt.xlabel("x (radians)", fontsize=12)
plt.ylabel("sin(x)", fontsize=12)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
