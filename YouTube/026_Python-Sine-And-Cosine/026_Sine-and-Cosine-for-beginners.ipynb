```python
import numpy as np  
import matplotlib.pyplot as plt  
import math
FONTSIZE = 10
```

```python
fig, ax = plt.subplots(figsize=(8, 8))  
ax.set_aspect('equal')  

# Draw bold x/y axes (main diameters)  
ax.axhline(0, color='red', lw=2, alpha=0.5)  # x-axis = cos(theta)  
ax.axvline(0, color='blue', lw=2, alpha=0.5)  # y-axis = sin(theta)  

# Draw unit circle  
theta = np.linspace(0, 2*np.pi, 100)  
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=1)  
ax.set_title("The Unit Circle: Your New Best Friend", fontsize=FONTSIZE+2)  
plt.show() 
```

```python
angles = [0, 90, 180, 270]  
colors = ['purple', 'orange', 'green', 'cyan']  

fig, ax = plt.subplots(figsize=(8, 8))  
ax.set_aspect('equal')  
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=1)  
ax.axhline(0, color='red', lw=2, alpha=0.3)  
ax.axvline(0, color='blue', lw=2, alpha=0.3)  

for deg, color in zip(angles, colors):  
    rad = np.deg2rad(deg)  
    x, y = np.cos(rad), np.sin(rad)  
    ax.scatter(x, y, s=200, color=color)  
    ax.text(x, y + 0.05, f'{deg}°', fontsize=FONTSIZE, color=color)  

ax.set_title("Cardinal Directions on the Circle", fontsize=FONTSIZE)  
plt.show()  
```

```python
angles = [30, 45, 60]  
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']  

fig, ax = plt.subplots(figsize=(8, 8))  
ax.set_aspect('equal')  
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=1)  
ax.axhline(0, color='red', lw=2, alpha=0.3)  
ax.axvline(0, color='blue', lw=2, alpha=0.3)  

for deg, color in zip(angles, colors):  
    rad = np.deg2rad(deg)  
    x, y = np.cos(rad), np.sin(rad)  
    ax.plot([0, x], [0, y], color=color, lw=2, alpha=0.5)  
    ax.scatter(x, y, s=200, color=color)  
    ax.text(x + 0.1, y + 0.1, f'θ = {deg}°\n({x:.2f}, {y:.2f})',  
            fontsize=FONTSIZE, color=color)  

ax.set_title("Your First Friends: 30°, 45°, 60°", fontsize=FONTSIZE+2)  
plt.show()  
```

```python
angles = [15, 165, 195, 345]  # α, 180-α, 180+α, -α  
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#9b59b6']  

fig, ax = plt.subplots(figsize=(8, 8))  
ax.set_aspect('equal')  
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=1)  
ax.axhline(0, color='red', lw=2, alpha=0.3)  
ax.axvline(0, color='blue', lw=2, alpha=0.3)  

for deg, color in zip(angles, colors):  
    rad = np.deg2rad(deg)  
    x, y = np.cos(rad), np.sin(rad)  
    ax.scatter(x, y, s=200, color=color)  
    ax.text(x + 0.1*np.sign(x), y + 0.1, f'θ = {deg}°',  
            fontsize=FONTSIZE, color=color, ha='left' if x > 0 else 'right')  

ax.set_title("Quadrant Party: Signs Change, Ratios Stay!", fontsize=FONTSIZE+2)  
plt.show()  
```

```python
theta_extended = np.linspace(0, 4*np.pi, 200)  # 0° to 720°  
x, y = np.cos(theta_extended), np.sin(theta_extended)  

fig, ax = plt.subplots(figsize=(12, 4))  
ax.plot(theta_extended, x, color='red', label='cos θ')  
ax.plot(theta_extended, y, color='blue', label='sin θ')  
ax.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi])  
ax.set_xticklabels(['0°', '180°', '360°', '540°', '720°'], fontsize=FONTSIZE)  
ax.set_title("Sine & Cosine: The Eternal Loop", fontsize=FONTSIZE+2)  
ax.grid(True)  
ax.legend()
plt.show()  
```

```python
fig, ax = plt.subplots(figsize=(8, 8))  
ax.set_aspect('equal')  
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=1)  
ax.axhline(0, color='red', lw=2, alpha=0.3)  
ax.axvline(0, color='blue', lw=2, alpha=0.3)  

# Example point  
angle = 25  
rad = np.deg2rad(angle)  
x, y = np.cos(rad), np.sin(rad)  
ax.scatter(x, y, s=200, color='purple')  
ax.plot([0, x], [0, y], 'purple', lw=2, alpha=0.3)  
ax.plot([x, x], [0, y], 'purple', linestyle='--')  
ax.plot([0, x], [0, 0], 'purple', linestyle='--')  

ax.text(0.5, 0.7,  
        r"$\cos^2\theta + \sin^2\theta = 1$" + "\n"  
        f"Example: ({x:.2f})² + ({y:.2f})² = 1",  
        fontsize=FONTSIZE, bbox=dict(facecolor='white', alpha=0.9))  

ax.set_title("Best Friend Rule: Always True!", fontsize=FONTSIZE+2)  
plt.show()  
```
