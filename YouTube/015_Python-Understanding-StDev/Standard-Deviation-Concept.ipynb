```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a random dataset
data = np.random.normal(0, 1, 1000)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Plot the data
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution')

# Fill the areas for each standard deviation segment
plt.fill_between(x, y, where=(x >= mean - std_dev) & (x <= mean + std_dev), color='blue', alpha=0.2, label='±1 SD (68.27%)')
plt.fill_between(x, y, where=(x >= mean - 2*std_dev) & (x <= mean + 2*std_dev), color='green', alpha=0.2, label='±2 SD (95.45%)')
plt.fill_between(x, y, where=(x >= mean - 3*std_dev) & (x <= mean + 3*std_dev), color='red', alpha=0.2, label='±3 SD (99.73%)')

plt.axvline(mean, color='black', linestyle='--', label='Mean')
plt.axvline(mean + std_dev, color='blue', linestyle='--')
plt.axvline(mean - std_dev, color='blue', linestyle='--')
plt.axvline(mean + 2*std_dev, color='green', linestyle='--')
plt.axvline(mean - 2*std_dev, color='green', linestyle='--')
plt.axvline(mean + 3*std_dev, color='red', linestyle='--')
plt.axvline(mean - 3*std_dev, color='red', linestyle='--')

plt.title('Standard Normal Distribution (Bell Curve)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

- 68.3% falls within 1 st. dev
- 95.4% falls within 2 st. dev
- 99.7% falls within 3 st. dev
