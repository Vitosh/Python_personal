```python
def start_state(n):
    return ">"*n+"_"+"<"*n

def goal_state(n):
    return "<"*n+"_"+">"*n

def print_path(path):
    for row in path:
        print(row)
```

```python
goal_state(3)
```

```python
def neighbours_research(state):
    """
      >*_ = _*>
      _*< = <*_
      >_  = _>
      _<  = _<
    """

    s = list(state)
    b = s.index("_")
    result = []

    # >*_._*>
    if b-2 >= 0 and s[b-2] == ">" and s[b-1] in "<>":
        t = s[:]
        t[b], t[b - 2] = t[b-2], t[b]
        result.append("".join(t))

    # _*<.<*_
    if b+2 < len(s) and s[b+2] == "<" and s[b+1] in "<>":
        t = s[:]
        t[b], t[b+2] = t[b+2], t[b]
        result.append("".join(t))

    # >_._>
    if b-1 >= 0 and s[b-1] == ">":
        t = s[:]
        t[b], t[b-1] = t[b-1], t[b]
        result.append("".join(t))

    # _<._<
    if b+1 < len(s) and s[b+1] == "<":
        t = s[:]
        t[b], t[b+1] = t[b+1], t[b]
        result.append("".join(t))
        
    return result
```

```python
neighbours_research(start_state(3))
```

```python
def solve_frogs_dfs(n):
    start = start_state(n)
    goal = goal_state(n)
    visited = set()
    path = []
    
    def dfs(state: str)-> bool:
        if state in visited:
            return False
        print(f"currently visited states - {len(visited)+1}")
        visited.add(state)
        path.append(state)
        print(f"✅added state:\n{state}")
        if state == goal:
            return True
        
        for next_state in neighbours_research(state):
            if dfs(next_state):
                return True
        
        
        print(f"❌deadend found:\n{path[len(path)-1]}")
        print(f"path so far, last is removed:")
        print_path(path)
        
        path.pop()
        
        return False
    
    dfs(start)
    print("✅✅✅✅✅✅")
    print_path(path)
    print("✅✅✅✅✅✅")

    return path
```

```python
solve_frogs_dfs(3)
```
